```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
```

```python
X = np.array([[2,3,1],[4,6,3],[5,4,2],[6,6,4],[8,7,5],[9,13,5],[12,23,8],[12,13,9],[15,13,14],[12,18,12]])
X
```

```python
y = np.array([3,4,9,14,22,34,46,57,67,79])
y
```

```python
fig, ax = plt.subplots(1,3,figsize = (12,3))

color=['blue','orange','green']

for i in range(len(ax)):
    ax[i].scatter(X[:,i], y, color=color[i])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X)
```

```python
fig, ax = plt.subplots(1,3,figsize = (12,3))

color=['blue','orange','green']

for i in range(len(ax)):
    ax[i].scatter(X_train[:,i], y, color=color[i])
```

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()
reg = reg.fit(X_train, y)
```

```python
w = reg.coef_
b= reg.intercept_
print(f'w is : {w} and b is : {b}')
```

```python
y_pred = reg.predict(X_train)
```

```python

fig,ax=plt.subplots(1,3,figsize=(12,3),sharey=True)


for i in range(len(ax)):
    ax[i].scatter(X_train[:,i],y,color=color[i], label = 'target')
    ax[i].scatter(X_train[:,i],y_pred,color='red', label = 'predict')
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y, y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y, y_pred))
```
